#MATPLOT LIBRARY

```python
import numpy as np
import matplotlib.pyplot as plt
student={"bob":45,"sam":78,"abc":89}
name=list(student.keys())
marks=list(student.values())
plt.bar(name,marks)
plt.title("student data")
plt.xlabel("name of students")
plt.ylabel("marks")
plt.show()
```
